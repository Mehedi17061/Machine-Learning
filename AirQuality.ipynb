```python
import requests
import csv
import json
from datetime import datetime

# Read the text file and print each line
with open('cityname.txt', 'r') as file:
    for line in file:
        # Parse the latitude, longitude, and city name from the line
        city, lat, lon = line.strip().split(',')
        print(f'{city}: ({lat}, {lon})')

        # Request the AQI code for the current city
        url = f"http://api.openweathermap.org/data/2.5/air_pollution?lat={lat}&lon={lon}&appid=d3be85c2db06fcae928ad942e17ae560"
        response = requests.get(url)
        # data = response.json()
        # print(aqi)
        # Check if the request was successful
        if response.status_code == 200:
            # Convert the response to a JSON object
            data = response.json()
            #print(data)

            # Extract the relevant data from the JSON object
            co = data['list'][0]['components']['co']
            no = data['list'][0]['components']['no']
            no2 = data['list'][0]['components']['no2']
            so2 = data['list'][0]['components']['so2']
            o3 = data['list'][0]['components']['o3']
            pm2_5 = data['list'][0]['components']['pm2_5']
            pm10 = data['list'][0]['components']['pm10']
            nh3 = data['list'][0]['components']['nh3']
            aqi = data['list'][0]['main']['aqi']
            print(aqi)


            current_date = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

            # Store the data in a list of dictionaries
            air_quality_data = [{"Date": current_date,"City":city, "PM2.5": pm2_5, "PM10": pm10, "SO2": so2, "NO2": no2,"CO":co,"NO":no,"O3":o3,"NH3":nh3,"AQI":aqi}]

            # Write the data to a CSV file
            with open("airQuality.csv", "a", newline="") as csvfile:
                fieldnames = ["Date","City","PM2.5", "PM10", "SO2", "NO2", "CO","NO","O3","NH3","AQI"]
                writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
                if csvfile.tell() == 0:
                    writer.writeheader()
                for row in air_quality_data:
                    writer.writerow(row)

            print("Data has been successfully written to air_quality.csv")
        else:
            print("Failed to retrieve data. Response status code:", response.status_code)

```
